```python
# for lambda value (regularizer) = 0.0

loss = []
regularizer = 0.0

for i in range(10):
    test_set = i+1

    # imoort the test data
    X_test = np.load("Fold" + str(test_set) + "/features.npy")
    y_test = np.load("Fold" + str(test_set) + "/target.npy")

    start_train = 1
    if test_set == 1:
        start_train = 2
    
    #import the train data
    X_train = np.load("Fold" + str(start_train) + "/features.npy")
    y_train = np.load("Fold" + str(start_train) + "/target.npy")

    for i in range(1,11,1):
        if i == start_train or i == test_set:
            continue
        x = np.load("Fold" + str(i) + "/features.npy")
        y = np.load("Fold" + str(i) + "/target.npy")
        X_train = np.concatenate((X_train,x),axis = 0)
        y_train = np.concatenate((y_train,y),axis = 0)
        
    #print(X_train.shape)
    #print(y_train.shape)
    # weights initialiation
    w = np.random.normal(size=(99, 1)).reshape((99,))  # X_train.shape[1] = 99
    b = 0
        
    # set the parameters
    max_iterations = 1000
    step_sz = 0.1
    
    # training part 
    for _ in range(max_iterations):
        
        prediction = X_train @ w + b
        grad_w = (X_train.T @ (prediction - y_train))/(X_train.shape[0]) + regularizer * w
        grad_b = np.mean(prediction - y_train)
        w = w - step_sz * grad_w
        b = b - step_sz * grad_b
    
        if np.linalg.norm(grad_w) + np.linalg.norm(grad_b) <= 0.001:
            print(_)
            break
    
    #print(w)
    loss.append(np.mean(np.square(X_test @ w+ b - y_test)))
        

print(np.mean(np.array(loss)))
```
